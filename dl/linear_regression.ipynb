```python
import torch
import random
```

```python
# Generate y=Wx+b data
def generate_data(W, b, num_examples):
    X = torch.normal(0, 1, (num_examples, len(W)))
    y = torch.matmul(X, W)+b
    y += torch.normal(0, 0.01, y.shape)
    return X, y
```

```python
features, labels = generate_data(torch.tensor([2, -3.4]).T, torch.tensor(4.2), 100)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i: min(i+batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 10

for X, y in data_iter(20, features, labels):
    print(X)
    print(y)
```

```python
W = torch.normal(0, 0.01, size=(1, 2), requires_grad=True).T
b = torch.zeros(1, requires_grad=True)
```

```python
def linear_model(X, w, b):
    return torch.matmul(X, w) + b
```

```python
def squared_loss(y_hat, y):
    return (y_hat-y.reshape(y_hat.shape))**2/2
```

```python
def sgd(params, lr, batch_size):
    with torch.no_grad():
        for para in params:
            para -= lr * para.grad/batch_size
            para.grad.zero_()
```

```python
lr = 0.03
num_epoches = 3
net = linear_model
loss = squared_loss

for epoch in range(num_epoches):
    for X, y in data_iter(batch_size, features, labels):
        l = squared_loss(linear_model(X, W, b), y)
        l.sum().backward()
```
